# Correcting words

```python
words = []
wordCount = 0
paraCount = 0
for i in range(0,len(paragraphList)):
    for word in paragraphList[i]:
        if word[0] in punctuations:
            word = word.replace(word[0],"")
            wordCount +=1
            words.append(word)
        elif word[-1] in punctuations:
            if word[-1] in punctuations:
                word = word.replace(word[-1]," ")
                wordCount +=1
                words.append(word)
                if word[-1] in punctuations:
                    word = word.replace(word[-1],"")
                    wordCount +=1
                    words.append(word)
        else:
            word = word
            wordCount +=1
            words.append(word)
    paraCount += 1
print(words)
print(paraCount)
print(wordCount)
```

```python
finalWords = []
for word in words:
    if "-" in word:
        print(word)
        wordSplit= word.split("-")
        for i in wordSplit:
            finalWords.append(i)
    elif "—" in word:
        print(word)
        wordSplit= word.split("—")
        for i in wordSplit:
            finalWords.append(i)
    else:
        finalWords.append(word)
```

```python
wordDict = {}
print(len(finalWords))
```

```python
counter = 1
for word in finalWords:
    if word in wordDict.keys():
        wordDict[word] = wordDict[word]+1
    if word not in wordDict.keys():
        wordDict[word] = counter
print(wordDict.values())
print(wordDict.keys())
```
